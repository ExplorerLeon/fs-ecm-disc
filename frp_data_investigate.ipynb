```python
import pandas as pd

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 500)
```

```python
# Path to the Excel file
file_path = "./input_data/Portfolio Monitoring (Dec 2023).xlsm"

# Read the Excel file into a pandas DataFrame, skipping the first row
df = pd.read_excel(file_path, sheet_name="Policies Outstanding DEC23", header=1)

# Display the DataFrame
df.head()  # Display the first few rows of the DataFrame
```

```python
df.tail(10) 
```

```python
df.columns
```

```python
df.columns = df.columns.str.strip()
```

```python
df.info()
```

```python
# Drop columns NaN values
df = df.dropna(axis=1, how="all")

df.head()
```

```python
df.info()
```

```python
df = df.dropna(how='all')
df.tail(10)
```

```python
df.info()
```

```python
df[df["Bond ID"].isna()]
```

```python
# Condition 1: Drop rows where both column A and column B have NaN values
condition_1 = df['Applicant/Borrower'].isna() & df['Bond ID'].isna()

# Condition 2: Drop rows where column A contains the string "Number of Deals" and column B has NaN value
condition_2 = (df['Applicant/Borrower'] == 'Number of Deals') & df['Bond ID'].isna()

# Combine the conditions using logical OR (|)
combined_condition = condition_1 | condition_2

# Extract the rows that do not meet the combined condition
filtered_df = df[~combined_condition]

# Display the filtered DataFrame
print("DataFrame after dropping rows:")
filtered_df.tail()
```
